```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Creating the dataset based on the table provided
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)
```

```python
# Separating the features (X) and the target variable (y)
X = df[['area']]  # Independent variable (area)
y = df['price']   # Dependent variable (price)
```

```python
# Initializing the linear regression model
model = LinearRegression()
```

```python
# Fitting the model with the data
model.fit(X, y)
```

```python
# Predicting the price for given areas (5000, 8000, 9000 sq ft)
areas_to_predict = pd.DataFrame([5000, 8000, 9000], columns=['area'])
predicted_prices = model.predict(areas_to_predict)
```

```python
# Displaying the predictions
for area, price in zip(areas_to_predict['area'], predicted_prices):
    print(f"Predicted price for a home with area {area} sq ft: ${price:.2f}")
```

```python
# Plotting the data and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(df['area'], df['price'], color='blue', label='Actual Prices')  # Scatter plot of actual data
plt.plot(df['area'], model.predict(X), color='red', label='Regression Line')  # Red line for regression
plt.scatter(areas_to_predict['area'], predicted_prices, color='green', label='Predicted Prices')  # Green points for predictions
plt.title('Area vs Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```
