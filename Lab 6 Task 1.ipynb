```python
# Importing necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Creating the dataset
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)
```

```python
# Separating features (X) and target variable (y)
X = df[['area']]
y = df['price']
```

```python
# Initializing and fitting the Linear Regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predicting prices for the given areas
predicted_prices = model.predict(X)
```

```python
# Adding predicted prices to the DataFrame for comparison
df['predicted_price'] = predicted_prices
```

```python
# Displaying the DataFrame
print(df)
```

```python
# Plotting the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(df['area'], df['price'], color='blue', label='Actual Prices')  # Scatter plot for actual prices
plt.plot(df['area'], df['predicted_price'], color='red', label='Regression Line')  # Regression line
plt.title('Area vs Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```
