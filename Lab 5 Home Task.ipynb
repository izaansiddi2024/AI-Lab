```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
# Step 1: Define the dataset
data = {
    "Genre": ["Action", "Comedy", "Drama", "Action", "Horror", "Drama", "Action",
              "Comedy", "Horror", "Action", "Comedy", "Drama", "Horror", "Drama"],
    "Watched": ["Yes", "No", "Yes", "No", "Yes", "No", "Yes", 
                "Yes", "No", "Yes", "Yes", "No", "Yes", "No"],
    "Classification": ["Good", "Bad", "Good", "Bad", "Good", "Bad", "Good", 
                       "Good", "Bad", "Good", "Good", "Bad", "Good", "Bad"]
}

# Step 2: Create a DataFrame
df = pd.DataFrame(data)
```

```python
# Step 3: Encode the categorical variables using LabelEncoder
label_encoder_genre = LabelEncoder()
label_encoder_watched = LabelEncoder()
label_encoder_classification = LabelEncoder()

df['Genre'] = label_encoder_genre.fit_transform(df['Genre'])  # Encode Genre
df['Watched'] = label_encoder_watched.fit_transform(df['Watched'])  # Encode Watched
df['Classification'] = label_encoder_classification.fit_transform(df['Classification'])  # Encode Classification
```

```python
# Step 4: Separate features (X) and target variable (y)
X = df[['Genre', 'Watched']]  # Features: Genre, Watched
y = df['Classification']  # Target: Classification
```

```python
# Step 5: Train a Decision Tree Classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python
# Step 6: Predict the classification for a new user entry
# Request new entry from the user
print("Enter the details for a new entry:")
new_genre = input("Genre (Action, Comedy, Drama, Horror): ").strip()
new_watched = input("Watched (Yes or No): ").strip()

# Encode the new entry using the same label encoders
new_genre_encoded = label_encoder_genre.transform([new_genre])[0]
new_watched_encoded = label_encoder_watched.transform([new_watched])[0]

# Convert new entry to a DataFrame with matching column names
new_entry = pd.DataFrame([[new_genre_encoded, new_watched_encoded]], columns=['Genre', 'Watched'])

# Predict the classification
predicted_classification = clf.predict(new_entry)[0]
predicted_classification_label = label_encoder_classification.inverse_transform([predicted_classification])[0]

# Output the predicted classification
print(f"\nPredicted Classification: {predicted_classification_label}")
```
